## Reduction operation: the sum of the numbers in the range [0, value)

```python
import time

def reduc_operation(a):
    """Compute the sum of the numbers in the range [0, a)."""
    x = 0
    for i in range(a):
        x += i
    return x

# Secuencial

value = 1000000

initialTime = time.time()
suma = reduc_operation(value)
finalTime = time.time()

print("Time taken by reduction operation:", (finalTime - initialTime), "seconds")

# Utilizando las operaciones mágicas de ipython
%timeit -r 2 reduc_operation(value)

print(f"\n \t Computing the sum of numbers in the range [0, value): {suma}\n")
```

a)En la siguiente celda de código del notebook vamos a utilizar el tipo de datos listas para realizar
dicha operación de reducción. Para ello, crea una lista de Python con la función range que contenga
10^6 elementos. Una vez creada, calcula el tiempo que tardas en sumar todos los elementos de la lista
de 2 formas diferentes: a) Usando un bucle for similar al código ya mostrado en la celda anterior
para ir sumando al resultado total elemento a elemento de la lista, y b) usando la función sum que
tienen las listas. Como verás, hay una diferencia importante a favor de usar la función sum.

```python

N = 10**6
lista_datos = list(range(N))

# Funcion para bucle for
def suma_manual(datos):
    "Suma los elementos de una lista usando un bucle for"
    total = 0
    for elemento in datos:
        total += elemento
    return total

tiempo_bucle = %timeit -r 2 -o -q suma_manual(lista_datos)

print("Tiempo tomado usando bucle for:", tiempo_bucle)
print(f"Resultado de la suma: {suma_manual(lista_datos)}\n")


# Metodo B
tiempo_sum = %timeit -r 2 -o -q sum(lista_datos)

print("Tiempo tomado usando sum():", tiempo_sum)
print(f"Resultado de la suma: {sum(lista_datos)}")
```

```python
b)En la segunda celda de código vamos a utilizar el tipo de datos array que nos proporciona el paquete
numpy. En primer lugar, pasa la lista creada en la celda anterior a un array de numpy. Y al igual que
antes, calcula el tiempo que tardas en sumar todos los elementos del array de 2 formas diferentes: a)
Usando un bucle for para ir sumando elemento a elemento del array al resultado total, y b) usando
la función sum que tiene el paquete numpy para los arrays.
```

```python
import numpy as np
N = 10**6
lista_datos = list(range(N))
array_datos = np.array(lista_datos)

#Funcion para sumar en bucle
def suma_manual_numpy(arr):
    s = 0
    for i in arr:
        s += i
    return s

# Usando bucle for sobre el Array .Apartado a)
tiempo_bucle_np = %timeit -r 2 -o -q suma_manual_numpy(array_datos)

print("Tiempo tomado usando bucle for en NumPy:", tiempo_bucle_np)
print(f"Resultado de la suma: {suma_manual_numpy(array_datos)}\n")


# Usando sum. Apartado b)
tiempo_numpy_sum = %timeit -r 2 -o -q np.sum(array_datos)

print("Tiempo tomado usando np.sum():", tiempo_numpy_sum)
print(f"Resultado de la suma: {np.sum(array_datos)}")
```
